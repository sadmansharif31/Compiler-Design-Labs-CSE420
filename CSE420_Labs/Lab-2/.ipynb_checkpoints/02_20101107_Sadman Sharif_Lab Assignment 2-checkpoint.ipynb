```python
def email(word):
    state=" "
    p=len(word)
    char=["_","+",".","-"]
    global end
    endd= ''
    top_level_domainn=['bd','com','org','net']
    
      
    for i in range(0,p):
        
        asc=ord(word[i])

        if state==" ":
            if 97<=asc<=122:
                state="1"
                continue
            else:
                return False

        if state=="1":
            if word[i] in char and word[i+1]==word[i]:
                return False
            if word[i] in char:
                state=" "

            elif 97<=asc<=122 or 48<=asc<=57:
                state="1"

            elif word[i]=="@":
                state="2"

            else:
                return False
            
            continue

        if state=="2":
            if 97<=asc<=122:
                state="3"
                continue
            else:
                return False

        if state=="3":
            
            if 97<=asc<=122:
                continue
            elif word[i]==".":
                state="4"
                continue
            else:
                return False
    
        if state=="4":
            j=i
            p=j+2
            while j<=p:
                endd+=word[j]
                j+=1
            if 97<=asc<=122:
                state="5"
                continue
            else:
                return False

        if state=="5":
            if endd in top_level_domainn:
                state = "Final"
                continue
            else:
                return False

    if state=="Final":
        return True
    else:
        return False

def web(word):
    state=" "
    global end 
    end=''
    top_level_domain=['com','net','org','edu','gov','mil','int']

    if word[0:4]=="www.":
        n=4
    else:
        n=0
    for i in range(n,len(word)):
        asc=ord(word[i])

        if state==" ":
            if 65 <=asc<= 90 or 48<=asc<=57 or 97<=asc<=122:
                state="1"
            else:
                return False

        if state=="1":
            if word[i]==".":
                state="2"
            elif not (65 <= asc<= 90 or 97 <= asc<= 122 or 48 <= asc<= 57):
                return False
            continue

        if state=="2":
            j=i
            p=j+2
            while j<=p:
                end+=word[j]
                j+=1
            if 65 <= asc<= 90 or 97 <= asc<= 122:
                state = "3"
                continue
            else:
                return False

        if state == "3":
            if end in top_level_domain:
                state="Final"
                continue
            else:
                return False

    if state =="Final":
        return True
    else:
        return False


#lst=[]
f=open("input.txt",'r') 
x=int(f.readline())

for i in range(0,x):
    temp=f.readline().strip("\n")
    w=web(temp)
    e=email(temp)
    
    if e is True:
        print("Email, "+str(i+1))
    
    elif w is True:
        print("Web, "+str(i+1))
    
    else:
        print("Neither an Email nor a Web address")

f.close()
```
